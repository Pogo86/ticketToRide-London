```python
import functions as fn
import pandas as pd
import matplotlib as plt
import seaborn as sns
import numpy as np
```

```python
# crteate a dataframe of the line data
linesdf = pd.read_csv('csv/lines.csv')
linesdf = linesdf.rename(columns={'Node 1': 'node1', 'Node 2':'node2','Line Color': 'colour', 'Line Length' : 'length'})

linesdf['id'] = linesdf.apply(fn.nameLines, axis=1)
linesdf.sort_values('id')
```

```python
# takes a copy of the dataframe swaps the columns node1 & node2 and joins the new and old dataframes
# this is done so lines can be used each way, eg waterloo to Globe Theatre and Globe Theatre to Waterloo
temp = linesdf
temp = temp.rename(columns={'node1':'x','node2':'y'})
temp =temp.rename(columns={'x':'node2','y':'node1'})
linesdf= pd.concat([linesdf, temp])
linesdf = linesdf.reset_index(drop=True)
```

```python
searchLocation = 'Globe Theatre'
linesdf.loc[(linesdf['node1'] == searchLocation) | (linesdf['node2'] == searchLocation)].sort_values('id')
```
